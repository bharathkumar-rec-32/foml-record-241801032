```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn import datasets
iris_data_loader = datasets.load_iris()
df = pd.DataFrame(iris_data_loader.data, columns=iris_data_loader.feature_names)
df["target"] = iris_data_loader.target
df.columns = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm', 'target']

df['Species'] = df['target'].map(lambda x: iris_data_loader.target_names[x]).astype('category')

print("Dataset Prepared for plotting:")
print(df.columns)
print(df.head())

df_Setosa = df[df['Species'] == 'setosa']
df_Versicolor = df[df['Species'] == 'versicolor']
df_Virginica = df[df['Species'] == 'virginica']

def plot_univariate(data1, data2, data3, feature, xlabel):
    plt.figure(figsize=(7,2));
    plt.scatter(data1[feature], np.zeros_like(data1[feature]), label='Setosa');
    plt.scatter(data2[feature], np.zeros_like(data2[feature]), label='Versicolor');
    plt.scatter(data3[feature], np.zeros_like(data3[feature]), label='Virginica');
    plt.xlabel(xlabel);
    plt.legend();
    plt.title(f'Univariate Plot of {xlabel}');
    plt.show()

plot_univariate(df_Setosa, df_Versicolor, df_Virginica, 'SepalWidthCm', 'Sepal Width (cm)')
plot_univariate(df_Setosa, df_Versicolor, df_Virginica, 'SepalLengthCm', 'Sepal Length (cm)')
plot_univariate(df_Setosa, df_Versicolor, df_Virginica, 'PetalWidthCm', 'Petal Width (cm)')
plot_univariate(df_Setosa, df_Versicolor, df_Virginica, 'PetalLengthCm', 'Petal Length (cm)')

sns.FacetGrid(df, hue='Species', height=5).map(plt.scatter, "SepalWidthCm", "PetalWidthCm").add_legend()
plt.title('Sepal Width vs Petal Width')
plt.show()

sns.FacetGrid(df, hue='Species', height=5).map(plt.scatter, "SepalLengthCm", "PetalLengthCm").add_legend()
plt.title('Sepal Length vs Petal Length')
plt.show()

sns.pairplot(df, hue="Species", height=2)
plt.suptitle('Pairplot of Iris Dataset Features', y=1.02)
plt.show()
```
